```python
import pandas as pd
```

```python
file_path ="/content/hospital-childbirth-1995-2023-supplementary-data-tables-en.xlsx"
```

```python
df = pd.read_excel(file_path, sheet_name="Table 1")

df.head()
```

```python
# Droping the first row which contains header descriptions
df_clean = df.drop(index=0).reset_index(drop=True)

# Renamed columns clearly
df_clean.columns = [
    "Fiscal_Year",
    "Province_Territory",
    "Age_Group",
    "Sex",
    "Number_of_Discharges",
    "Total_Length_of_Stay_Days",
    "Average_Length_of_Stay_Days"
]

# Remove rows with notes or missing provinces
df_clean = df_clean[~df_clean["Province_Territory"].isna()]

# Convert numeric columns
numeric_cols = [
    "Number_of_Discharges",
    "Total_Length_of_Stay_Days",
    "Average_Length_of_Stay_Days"
]

for col in numeric_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")

df_clean.head()
```

```python
# Remove rows with notes or missing provinces
df_clean = df_clean[~df_clean["Province_Territory"].isna()]
```

```python
# Convert numeric columns
numeric_cols = [
    "Number_of_Discharges",
    "Total_Length_of_Stay_Days",
    "Average_Length_of_Stay_Days"
]

for col in numeric_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")

df_clean.head()
```

```python
df_clean.isna().sum()
```

```python
df_cleaned = df_clean.dropna(subset=numeric_cols)
```

```python
df_cleaned.isna().sum()
```

```python
# Remove footnotes rows if any appear later
df_cleaned = df_cleaned[~df_cleaned["Age_Group"].astype(str).str.contains("Note", case=False, na=False)]

# Save cleaned dataset
output_path = "/content/canada_hospital_inpatient_clean_1995_2024.csv"
df_cleaned.to_csv(output_path, index=False)

output_path
```
